## Feature Selection

```python
# Find out the important features from this dataset

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
X = df.iloc[:,0:13]
y = df.iloc[:,-1]    #target column i.e price range

X.head()
```

```python
y = np.round(df['Price'])

y.head()
```

```python
# Applying SelectKBest class to extract top 10 best features from the Dataset

bestfeatures = SelectKBest(score_func=chi2, k=5)
fit = bestfeatures.fit(X,y)
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
```

```python
# Concatenating two dataframes for better graphical visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  # naming the dataframe columns
```

```python
featureScores
```

```python
# Displaying the Best Five Features

print(featureScores.nlargest(5,'Score'))
```

### Feature Importance

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(X,y)
```

```python
# Using in-built class feature_importances of tree based classifiers 

print(model.feature_importances_)
```

```python
# Plotting graph of Features importances for better graphical visualization

feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

## Model Fitting 

```python
X = df.iloc[:,0:13]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_train)
```

```python
print("Training Accuracy:",model.score(X_train, y_train) * 100)
```

```python
print("Testing Accuracy:",model.score(X_test,y_test) * 100)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print("Model Accuracy:", r2_score(y,model.predict(X)) * 100)
```

```python
plt.scatter(y_train, y_pred)
plt.xlabel("Prices")
plt.ylabel("Predicted prices")
plt.title("Prices vs Predicted prices")
plt.show()
```

```python
# Checking for the Residuals

plt.scatter(y_pred,y_train-y_pred)
plt.title("Predicted vs residuals")
plt.xlabel("Predicted")
plt.ylabel("Residuals")
plt.show()
```

```python
# Checking for the Normality of Errors

sns.distplot(y_train-y_pred)
plt.title("Histogram of Residuals")
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.show()
```

## Random Forest Regressor

```python
X = df.iloc[:,[-1,5,10,4,9]]
y = df.iloc[:,[-1]]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor()
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_train)
```

```python
print("Training Accuracy:",reg.score(X_train, y_train) * 100)
```

```python
print("Testing Accuracy:",reg.score(X_test, y_test) * 100)
```

```python
# Visualizing the differences between Actual prices and Predicted values

plt.scatter(y_train, y_pred)
plt.xlabel("Prices")
plt.ylabel("Predicted prices")
plt.title("Prices vs Predicted prices")
plt.show()
```
